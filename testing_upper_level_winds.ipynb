# Upper levels winds as a predictor #

The upper level winds (e.g. U850) appear to be more skillful and less noisy in the hindcast data, so we want to establish the potential for using these to predict offshore wind capacity factors. To first demonstrate that zonal wind is sound, we can plot the correlations between the observed 850U winds and the observed 100m winds. Then we can look at the observed correlations between 850U and offshore wind CF.'s and 100m winds and offshore wind CF.'s respectively. To summarise we are plotting:

* Observed (ERA5) correlations(850U, 100m wind speed)
* Observed (ERA5 + UREAD-ERA5) correlations (850U, offshore wind CF.'s)
* Observed (ERA5 + UREAD-ERA5) correlations (100m wind speed, offshore wind CF.'s)
